```python
import pandas as pd
!pip install pyarrow
from matplotlib import pyplot as plt
```

```python
data = pd.read_parquet("./2021-Project-74/results/grid_search/45/ts_model-state.parquet")
```

```python
data['max_interest'] = data['interest'].apply(max)
```

```python
import numpy as np
from scipy import interpolate
```

```python
X = data['p']
Y = data['w']
Z = data['max_interest']

fig = plt.figure(figsize=(6,6))
ax = fig.add_subplot(projection='3d')
ax.scatter(Y, X, Z)
```

```python
data_1999 = data.where(data['round'] == 1999)

X = data_1999['p']
Y = data_1999['w']
Z = data_1999['max_interest']

fig = plt.figure(figsize=(6,6))
ax = fig.add_subplot(projection='3d')
ax.scatter(Y, X, Z)
```

```python
data_1999 = data.where(data['round'] == 1999)

cols = ['p', 'w', 'max_interest']
xyz = data_1999[cols].groupby(['p','w']).mean(['max_interest'])
xyz = xyz.reset_index()
xyz.columns =  ['p', 'w', 'max_interest']
xyz = xyz.pivot('p', 'w')

X=xyz.columns.levels[1].values
Y=xyz.index.values
Z=xyz.values
Xi,Yi = np.meshgrid(X, Y)
plt.contourf(Xi, Yi, Z, 10, alpha=0.7, cmap=plt.cm.jet);
```
